```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data_path = '../../../data/LungsCheck/LungsCheck_V1/train'
test_data_path = '../../../data/LungsCheck/LungsCheck_V1/test'
val_data_path = '../../../data/LungsCheck/LungsCheck_V1/val'
```

```python
covid_train_path = '../../../data/LungsCheck/LungsCheck_V1/train/COVID'
covid_test_path = '../../../data/LungsCheck/LungsCheck_V1/test/COVID'
covid_val_path = '../../../data/LungsCheck/LungsCheck_V1/val/COVID'

image_count_covid_train = 0
for address, dirs, files in os.walk(covid_train_path):
    for name in files:
        image_count_covid_train += 1
print(f'Number of X-ray train images of the lungs of people with COVID-19 - {image_count_covid_train}')

image_count_covid_test = 0
for address, dirs, files in os.walk(covid_test_path):
    for name in files:
        image_count_covid_test += 1
print(f'Number of X-ray test images of the lungs of people with COVID-19 - {image_count_covid_test}')

image_count_covid_val = 0
for address, dirs, files in os.walk(covid_val_path):
    for name in files:
        image_count_covid_val += 1
print(f'Number of X-ray val images of the lungs of people with COVID-19 - {image_count_covid_val}')
```

```python
normal_train_path = '../../../data/LungsCheck/LungsCheck_V1/train/NORMAL'
normal_test_path = '../../../data/LungsCheck/LungsCheck_V1/test/NORMAL'
normal_val_path = '../../../data/LungsCheck/LungsCheck_V1/val/NORMAL'

image_count_normal_train = 0
for address, dirs, files in os.walk(normal_train_path):
    for name in files:
        image_count_normal_train += 1
print(f'Number of X-ray train images of the normal lungs - {image_count_normal_train}')

image_count_normal_test = 0
for address, dirs, files in os.walk(normal_test_path):
    for name in files:
        image_count_normal_test += 1
print(f'Number of X-ray test images of the normal lungs - {image_count_normal_test}')

image_count_normal_val = 0
for address, dirs, files in os.walk(normal_val_path):
    for name in files:
        image_count_normal_val += 1
print(f'Number of X-ray val images of the normal lungs - {image_count_normal_val}')
```

```python
tuberculosis_train_path = '../../../data/LungsCheck/LungsCheck_V1/train/TUBERCULOSIS'
tuberculosis_test_path = '../../../data/LungsCheck/LungsCheck_V1/test/TUBERCULOSIS'
tuberculosis_val_path = '../../../data/LungsCheck/LungsCheck_V1/val/TUBERCULOSIS'

image_count_tuberculosis_train = 0
for address, dirs, files in os.walk(tuberculosis_train_path):
    for name in files:
        image_count_tuberculosis_train += 1
print(f'Number of X-ray train images of the normal lungs of people with tuberculosis - {image_count_tuberculosis_train}')

image_count_tuberculosis_test = 0
for address, dirs, files in os.walk(tuberculosis_test_path):
    for name in files:
        image_count_tuberculosis_test += 1
print(f'Number of X-ray test images of the normal lungs of people with tuberculosis - {image_count_tuberculosis_test}')

image_count_tuberculosis_val = 0
for address, dirs, files in os.walk(tuberculosis_val_path):
    for name in files:
        image_count_tuberculosis_val += 1
print(f'Number of X-ray val images of the nromal lungs of people with tuberculosis - {image_count_tuberculosis_val}')
```

```python
pneumonia_train_path = '../../../data/LungsCheck/LungsCheck_V1/train/PNEUMONIA'
pneumonia_test_path = '../../../data/LungsCheck/LungsCheck_V1/test/PNEUMONIA'
pneumonia_val_path = '../../../data/LungsCheck/LungsCheck_V1/val/PNEUMONIA'

image_count_pneumonia_train = 0
for address, dirs, files in os.walk(pneumonia_train_path):
    for name in files:
        image_count_pneumonia_train += 1
print(f'Number of X-ray train images of the lungs of people with pneumonia - {image_count_pneumonia_train}')

image_count_pneumonia_test = 0
for address, dirs, files in os.walk(pneumonia_test_path):
    for name in files:
        image_count_pneumonia_test += 1
print(f'Number of X-ray test images of the normal lungs of people with pneumonia - {image_count_pneumonia_test}')

image_count_pneumonia_val = 0
for address, dirs, files in os.walk(pneumonia_val_path):
    for name in files:
        image_count_pneumonia_val += 1
print(f'Number of X-ray val images of the nromal lungs of people with pneumonia - {image_count_pneumonia_val}')
```

```python
graph_data_train = {'COVID-19': image_count_covid_train,
                    'NORMAL LUNGS': image_count_normal_train,
                    'TUBERCULOSIS': image_count_tuberculosis_train,
                    'PNEUMONIA': image_count_pneumonia_train}

train_df = pd.DataFrame(graph_data_train, index = [0])
train_df.plot(kind='bar')
plt.xticks([0],['Lungs Images train'])
plt.xticks(rotation=0)
plt.show()
```
